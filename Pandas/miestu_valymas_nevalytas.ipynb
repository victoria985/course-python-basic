```python
def remove_space(string):
    if ' ' in string:
        return int(string.replace(' ', ''))
    return int(string)
        
            
    
```

```python
def remove_asterix(string):
    if type(string) == str:
        if '*' in string:
            return int(string.replace('*', ''))
        return int(string)
    return string
```
