```python
def fix_col(x):
    if type(x) == str:
        if ' ' in x:
            new_number = int(x.replace(' ', ''))
            return new_number
        elif '*' in x:
            new_number = int(x.replace('*', ''))
            return new_number
        elif '?' in x:
            return 0
        return int(x)
    return x
    
```

```python
miestai_copy.head()
```

```python
for col in miestai.columns:
    miestai_copy[col] = miestai_copy[col].apply(fix_col)
```
